```python
import os
os.makedirs("downloaded_files", exist_ok=True)
```

```python
import gdown, re

def fix_drive_link(link):
    m = re.search(r'id=([a-zA-Z0-9_-]+)', link)
    if m:
        return f"https://drive.google.com/file/d/{m.group(1)}/view"
    return link

with open("linkcpns.txt") as f:
    links = [l.strip() for l in f if l.strip()]

print("Total link:", len(links))

for i, link in enumerate(links, 1):
    print(f"[{i}/{len(links)}] downloading...")
    gdown.download(
        fix_drive_link(link),
        output="downloaded_files",
        quiet=False
    )
```
